# 3D scatterplot

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.style.use('_mpl-gallery')
```

```python
# let us make data
np.random.seed(19680801)
n = 100

rng = np.random.default_rng()

xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# let us plot the 3D scatterplot
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[], yticklabels=[], zticklabels=[])

plt.show()
```

```python
print(zs)
```

# 3D surface plot

```python
from matplotlib import cm
```

```python
plt.style.use('_mpl-gallery')
```

```python
# let us make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)

X, Y = np.meshgrid(X, Y)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# let us plot the surface
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})

ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[], yticklabels=[], zticklabels=[])

plt.show()
```

```python
print(Z)
```

# Triangular 3D surfaces

```python
plt.style.use('_mpl-gallery')

n_radii = 8
n_angles = 36

# let us make radii and angles spaces
radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)[..., np.newaxis]

# let us convert polar (radii, angles,) cords to cartesian (x, y) coords
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())
z = np.sin(-x*y)

# let us plot the trisurf
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})

ax.plot_trisurf(x, y, z, vmin=z.min() * 2, cmap=cm.Blues)
ax.set(xticklabels=[], yticklabels=[], zticklabels=[])

plt.show()
```

```python
print(z)
```

# 3D Voxel / volumetric plot

```python
plt.style.use('_mpl-gallery')
# let us prepare some coordinates
x, y, z = np.indices((8, 8, 8))

# now let us draw cuboids in the top left and bottom right corners
cube1 = (x < 3) & (y < 3) & (z < 3)
cube2 = (x >= 5) & (y >=5) & (z >= 5)

# combine the objects into a single boolean array
voxelarray = cube1 | cube2

# it is time plot 3D Voxel
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})

ax.voxels(voxelarray, edgecolors='k')

ax.set(xticklabels=[], yticklabels=[], zticklabels=[])

plt.show()
```

#### voxels example2

```python
# set the figure size
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# let us take random data points between 0 and 1
data = np.random.choice([0, 1], size=(5, 7, 9), p=[0.65, 0.35])

# let us create a new figure
fig = plt.figure()

# Axis with 3D projection
ax = fig.add_subplot(111, projection='3d')

# plot the voxels
ax.voxels(data, edgecolor='k', facecolors='red')

# display the plot
plt.show()
```

# 3D wireframe plot

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
plt.style.use('_mpl-gallery')
```

```python
# let us make data
X, Y, Z = axes3d.get_test_data(0.05)

# let us plot 3D wireframe 
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)


ax.set(xticklabels=[], yticklabels=[], zticklabels=[])

plt.show()
```
