# hist(x)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('_mpl-gallery')

# let us make our data
np.random.seed(1)

X = 4 + np.random.normal(0, 1.5, 200)

# let us ploat
fig, ax = plt.subplots()
ax.hist(X, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```

```python
print(X)
```

# errorbar(x, y, yerr, xerr)

```python
plt.style.use('_mpl-gallery')

# make data:
np.random.seed(1)

x = [2, 4, 6]
y = [3.6, 5, 4.2]
yerr = [0.9, 1.2, 0.5]


# let us plot
fig, ax = plt.subplots()

ax.errorbar(x, y, yerr, fmt='o', linewidth=2, capsize=6)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

# violinplot(D)

```python
plt.style.use('_mpl-gallery')

# make data:
np.random.seed(10)

D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

# let us plot violinplot
fig, ax = plt.subplots()
vp = ax.violinplot(D, [2, 4, 6], widths=2, showmeans=False, showmedians=False, showextrema=False)


# do the styling:
for body in vp['bodies']:
    body.set_alpha(0.9)
    
ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

# hist2d(x, y)

```python
plt.style.use('_mpl-gallery-nogrid')

# let us make data: correlated + noise
np.random.seed(1)

x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3


# plot hist2d
fig, ax = plt.subplots()
ax.hist2d(x, y, bins=(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)))

ax.set(xlim=(-2, 2), ylim=(-3, 3))

plt.show()
```

```python
print(y)
```

```python
x = np.random.normal(size=50000)
y = x * 3 + np.random.normal(size=50000)

#Big bans
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.jet)
plt.show()

#Small bans
plt.hist2d(x, y, bins=(300, 300), cmap=plt.cm.jet)
plt.show()

# unequal bins
plt.hist2d(x, y, bins=(300, 30), cmap=plt.cm.jet)
plt.show()
```

# hexbin(x, y, C)

```python
plt.style.use('_mpl-gallery-nogrid')

# let us make data: correlated + noise
np.random.seed(1)

x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

# let us plot hexbin
fig, ax = plt.subplots()
ax.hexbin(x, y, gridsize=20)

ax.set(xlim=(-2, 2), ylim=(-3, 3))

plt.show()
```

```python
x = np.random.normal(size=50000)
y = x * 3 + np.random.normal(size=50000)


# make the plot
plt.hexbin(x, y, gridsize=(15, 15))
plt.show()

# we can control the size of the bins:
plt.hexbin(x, y, gridsize=(150, 150))
plt.show()
```

# pie(x)

```python
plt.style.use('_mpl-gallery-nogrid')

# let us make data
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# let us plot the pie
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4), wedgeprops={"linewidth":1, "edgecolor":"white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.axis('equal')

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]

ax.pie(students, labels = langs, autopct='%1.2f%%')
plt.show()
```
